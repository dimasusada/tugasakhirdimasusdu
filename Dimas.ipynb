```python
import json  # mengambil fungsi untuk menyimpan dan membaca data dari file
import time  # OPTIMASI: digunakan untuk mengukur waktu eksekusi

# d = data warteg (bahan/menu), disingkat untuk penggunaan dalam loop
data_warteg = []  # list kosong untuk menyimpan data

# simpan data ke file JSON
def save_data():
    with open("warteg_data.json", "w") as f:  # buka file dengan mode tulis
        json.dump(data_warteg, f)  # simpan data list ke dalam file JSON

# muat data dari file JSON
def load_data():
    try:
        with open("warteg_data.json", "r") as f:  # buka file dengan mode baca
            return json.load(f)  # baca isi file dan ubah jadi list Python
    except:
        return []  # kalau file belum ada atau rusak, kembalikan list kosong

# mencari data berdasarkan kode
def search_item(kode):
    for d in data_warteg:  # loop semua data
        if d['kode'] == kode:  # jika kode cocok
            return d  # kembalikan data yang ditemukan
    return None  # jika tidak ada, kembalikan None
# OPTIMASI: cocok untuk data kecil. Untuk banyak data bisa gunakan dict.

# tambah data baru ke list
def tambah_item():
    kode = input("Kode bahan/menu: ")  # minta input kode
    if search_item(kode):  # cek jika kode sudah ada
        print("Kode ini sudah terdaftar.")  # tampilkan pesan error
        return  # kembali ke menu utama
    nama = input("Nama bahan/menu: ")  # input nama item
    jenis = input("Jenis (bahan/menu): ").lower()  # input jenis dan ubah ke huruf kecil
    if jenis not in ['bahan', 'menu']:  # validasi jenis
        print("Jenis harus 'bahan' atau 'menu'.")  # tampilkan error jika salah
        return
    stok = int(input("Jumlah stok/porsi: "))  # input stok (integer)
    data_warteg.append({  # tambahkan data ke list
        "kode": kode,
        "nama": nama,
        "jenis": jenis,
        "stok": stok
    })
    print("Data berhasil ditambahkan.\n")  # konfirmasi penambahan data

# tampilkan semua data
def lihat_semua():
    if not data_warteg:  # cek jika list masih kosong
        print("Belum ada data.")  # tampilkan pesan
        return
    start = time.time()  # OPTIMASI: catat waktu mulai
    for d in data_warteg:  # loop setiap data
        print(f"{d['kode']} - {d['nama']} ({d['jenis']}, Stok: {d['stok']})")
    end = time.time()  # catat waktu selesai
    print(f"Waktu tampilkan data: {end - start:.6f} detik\n")  # tampilkan durasi

# mengubah stok data
def edit_stok():
    kode = input("Masukkan kode bahan/menu: ")  # input kode item
    d = search_item(kode)  # cari item berdasarkan kode
    if d:  # jika ditemukan
        d['stok'] = int(input("Masukkan jumlah stok/porsi baru: "))  # ubah stok
        print("Stok diperbarui.\n")  # konfirmasi
    else:  # jika tidak ditemukan
        print("Data tidak ditemukan.\n")  # tampilkan pesan

# menghapus item dari list
def hapus_item():
    kode = input("Masukkan kode bahan/menu: ")  # input kode
    d = search_item(kode)  # cari item
    if d:  # jika ditemukan
        data_warteg.remove(d)  # hapus dari list
        print("Data berhasil dihapus.\n")  # konfirmasi
    else:  # jika tidak ada
        print("Data tidak ditemukan.\n")  # pesan error

# laporan stok (dari stok terbanyak)
def laporan_stok():
    if not data_warteg:  # jika list kosong
        print("Belum ada data.")  # tampilkan pesan
        return
    print("=== Laporan Stok Tertinggi ===")
    start = time.time()  # OPTIMASI: catat waktu mulai
    urut = sorted(data_warteg, key=lambda x: x['stok'], reverse=True)  # urutkan berdasarkan stok dari besar ke kecil
    for d in urut:  # tampilkan satu per satu
        print(f"{d['nama']} ({d['jenis']}) - Stok: {d['stok']}")
    end = time.time()  # waktu selesai
    print(f"Waktu proses laporan: {end - start:.6f} detik\n")  # tampilkan waktu proses

# """" PROGRAM UTAMA DAN ALUR KERJA """"
data_warteg = load_data()  # load data saat program dimulai

# loop menu utama
while True:
    print("=== Menu Inventori Warteg ===")
    print("1. Tambah Bahan/Menu")
    print("2. Lihat Semua Data")
    print("3. Edit Stok")
    print("4. Hapus Data")
    print("5. Laporan Stok Tertinggi")
    print("6. Simpan & Keluar")

    pilihan = input("Pilih menu (1-6): ")  # input pilihan user
    print()

    # percabangan berdasarkan pilihan user
    if pilihan == '1': tambah_item()
    elif pilihan == '2': lihat_semua()
    elif pilihan == '3': edit_stok()
    elif pilihan == '4': hapus_item()
    elif pilihan == '5': laporan_stok()
    elif pilihan == '6':  # jika keluar
        save_data()  # simpan data sebelum keluar
        print("Data disimpan. Keluar.")  # pesan konfirmasi
        break  # keluar dari loop
    else:
        print("Pilihan tidak valid.\n")  # jika input bukan 1-6
```
